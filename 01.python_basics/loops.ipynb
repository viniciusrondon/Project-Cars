# Loops

# For

```python
clients = [
    'John',
    'Jane',
    'Bob',
    'Alice',
]

def send_email(client):
    print(f"Email sent to {client}")
```

```python
for client in clients:
    send_email(client)
```

```python
for i, client in enumerate(clients):
    print(i, client)
    send_email(client)
```

```python
for i, client in enumerate(clients):
    if i == 2:
        break
    print(i, client)
    send_email(client)
```

```python
for i, client in enumerate(clients):
    if i == 2:
        continue
    print(i, client)
    send_email(client)
```

# while

```python
number = 0 

while number < 5:
    print(number)
    number += 1
```
