```python
# example of a class

class Cellphone:
    def __init__(self, brand, model, color,has_camera,battery_level):
        self.brand = brand
        self.model = model
        self.color = color
        self.has_camera = True
        self.battery_level = 100

    def make_call(self, phone_number):
        print(f"Calling {phone_number}...")

    def send_sms(self, phone_number, message):
        print(f"Sending SMS to {phone_number}: {message}")
    
    def charge(self, hours):
        print(f"Charging for {hours} hours...")
```

```python
my_cellphone = Cellphone("Samsung", "Galaxy S21", "Black", True, 100)

print(my_cellphone.brand)
```

```python
my_cellphone.make_call("123-456-7890")
```

## Example 2 - heritage

```python
class Car: 
    def __init__(self, brand, model, color, has_sunroof, wheels, doors):
        self.brand = brand
        self.model = model
        self.color = color
        self.has_sunroof = has_sunroof
        self.wheels = wheels
        self.doors = doors
        self.speed = 0

    def start(self):
        print("Vroom vroom!")
    
    def acelerating(self, time):
        self.speed = self.speed + 1 * time
        print(f"The car is acelerating at {self.speed} km/h.")
    
    def braking(self, time):
        self.speed = self.speed - 1 * time
        if self.speed < 0:
            self.speed = 0
        print(f"The car is braking at {self.speed} km/h.")

    def honk(self):
        print("Beep beep!")
```

```python
my_car = Car("Toyota", "Corolla", "Blue", True, 4, 4)
print(my_car.color)
print(my_car.brand)
my_car.start()
my_car.acelerating(10)
my_car.braking(5)
my_car.honk()
```

```python
class Civic(Car):
    def __init__(self, color, has_sunroof, wheels, doors):
        super().__init__("Honda", "Civic", color, has_sunroof, wheels, doors)
```

```python
my_civic = Civic("Blue", True, 4, 4)
print(my_civic.brand)
print(my_civic.model)
print(my_civic.color)
print(my_civic.has_sunroof)
print(my_civic.wheels)
print(my_civic.doors)
my_civic.start()
my_civic.acelerating(10)
my_civic.braking(5)
my_civic.honk()
```

## Example 3 - polymorphism

```python
class Animal:
    
    def make_sound(self):
        print("Animal makes a sound")

class Dog(Animal):
    def make_sound(self):
        print("Dog barks")

class Cat(Animal):
    def make_sound(self):
        print("Cat meows")

dog = Dog()
dog.make_sound()

cat = Cat()
cat.make_sound()
```

## Example 4 - polymorphism

```python
class Form:
    def area(self):
        pass
    
    def perimeter(self):
        pass

class Square(Form):
    def __init__(self, side: int) -> None:
        self.side = side
    
    def area(self):
        return self.side ** 2
    
    def perimeter(self):
        return 4 * self.side

my_square = Square(5)
print(my_square.area())
print(my_square.perimeter())
```
