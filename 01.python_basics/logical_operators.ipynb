# Logical operators

## not

invert the value of the operand

```python
# not
a = True
b = False

print(not a)
print(not b)
```

## and

Is True when both operands are true

```python
# and

print(True and True)
print(True & False)
print(False and True)
print(False and False)
```

```python
if a and b:
    print("Met the condition")
else:
    print("Failed the condition")
```

## or

Is True when at least one of the operands is true

```python
# or 

print(True or True)
print(True or False)
print(False or True)
print(False or False)
```

## Xor 
Is True when only one of the operands is true

```python
# xor

print(True ^ True)
print(True ^ False)
print(False ^ True)
print(False ^ False)
```

```python
name = "Vinicius"
age = 25
weight = 110

if (name == "Vinicius" or age == 25) and weight == 90:
    print("You are Vinicius")
else:
    print("You are not Vinicius")
```
