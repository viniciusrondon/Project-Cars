# Strings

strings in python are considered as a sequence of characters such as lists so, we can use indexing and slicing to access the characters in a string.

```python
text = "Hello World"
print(text)
```

```python
text[0]
```

```python
text[6:11]
```

```python
len(text)
```

```python
text.count("l") 
```

```python
text.count("l",0,5)
```

```python
text.find("World")
```

```python
text.upper()
```

```python
text.lower()
```

```python
text.capitalize()
```

```python
text.title()
```

```python
text.split(" ")
```

```python
word_list = text.split(" ")
"_".join(word_list)
```
